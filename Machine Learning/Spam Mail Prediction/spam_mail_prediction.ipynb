label\

text\
This column contains the actual content of the email messages.

```python
# importing libraires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score,mean_squared_error,r2_score
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df=pd.read_csv('email.csv')
df.head()
```

```python
df.info()
```

```python
df.shape
```

```python
# replacing the null values with null strings
mail_df=df.where(pd.notnull(df), '')
mail_df.head()
```

```python
# encoding the categorical columns
le=LabelEncoder()
mail_df['Category']=le.fit_transform(mail_df['Category'])
mail_df.head()
```

```python
# seperating the text and labels
X=mail_df['Message']
y=mail_df['Category']
```

```python
print(X)
```

```python
# SPLITTING the training and test
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# transform the text data into vectors 
vectorizer=TfidfVectorizer(min_df=1,stop_words='english',lowercase=True)
X_train=vectorizer.fit_transform(X_train)
X_test=vectorizer.transform(X_test)

# convertingb thr ytrain and y test values as integer
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(X_train)
```

```python
# training the model
model=LogisticRegression()
model.fit(X_train,y_train)
```

```python
# model evaluation
y_pred=model.predict(X_test)
print('accuracy score:',accuracy_score(y_test,y_pred))
print('mean squared error:',mean_squared_error(y_test,y_pred))
print('r2 score:',r2_score(y_test,y_pred))
```

```python
# predition on training data
y_train_pred=model.predict(X_train)
print('accuracy score:',accuracy_score(y_train,y_train_pred))
print('mean squared error:',mean_squared_error(y_train,y_train_pred))
print('r2 score:',r2_score(y_train,y_train_pred))
```

```python
# building a preditive system
input_mail=["Hi Sarah, your appointment with Dr. Smith is confirmed for tomorrow at 3 PM. Please reply YES to confirm or call us if you need to reschedule."]
input_mail=vectorizer.transform(input_mail)
print(model.predict(input_mail))
print(model.predict_proba(input_mail))
```
